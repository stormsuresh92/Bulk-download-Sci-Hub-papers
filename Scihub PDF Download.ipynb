```python
import requests
from bs4 import BeautifulSoup
import time
from tqdm import tqdm
import wget
```

```python
#url = 'https://sci-hub.se/https://doi.org/10.1248/cpb.40.1338'

header = {
    "Host":"sci-hub.se",
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:122.0) Gecko/20100101 Firefox/122.0",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.5",
    "Accept-Encoding": "gzip, deflate, br",
    "Connection": "keep-alive",
    "retry-After": "3600",
    "Upgrade-Insecure-Requests": "1",
    "Sec-Fetch-Dest": "document",
    "Sec-Fetch-Mode": "navigate",
    "Sec-Fetch-Site": "cross-site"
}
```

```python
f = open('doilist.txt', "r")
dois = f.readlines()
for doi in tqdm(dois):
  res = requests.get("https://sci-hub.se/"+doi.strip(), headers=header)
  time.sleep(3)
  if res.status_code==200:
    try:
      soup = BeautifulSoup(res.content, 'html.parser')
      pdftag = soup.find("embed")
      pdfurl = pdftag.get('src').replace("#navpanes=0&view=FitH", "")
      if pdfurl.startswith("//"):
        pdfurl = "https:"+pdfurl
      elif pdfurl.startswith("/"):
        pdfurl = "https://sci-hub.se"+pdfurl
      else:
        print(pdfurl)
      #Download pdf files using wget library
      wget.download(pdfurl)

      #saving DOI along with PDF url in csv format
      f = open("pdfurl.csv", 'a')
      f.write(doi.strip()+","+pdfurl+"\n")
    except:
      f = open("pdfurl.csv", 'a')
      f.write(doi.strip()+","+"PDF not found"+"\n")

  else:
    print("Pdf not found")
```
